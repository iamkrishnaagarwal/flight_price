```python
def sub(s):
    if (s =="New Delhi"):
        return "Delhi"
    else:
        return s
train["Destination"] = train["Destination"].apply(sub)
test["Destination"] = test["Destination"].apply(sub)
```

```python
def sub(s):
    if (s =="Delhi"):
        return 0
    elif (s=="Kolkata"):
        return 1
    elif (s=="Banglore"):
        return 2
    elif (s=="Cochin"):
        return 3
    elif (s=="Hyderabad"):
        return 4
    elif (s=="Chennai"):
        return 5
    elif (s=="Mumbai"):
        return 6
train["Source"] = train["Source"].apply(sub)
train["Destination"] = train["Destination"].apply(sub)
test["Source"] = test["Source"].apply(sub)
test["Destination"] = test["Destination"].apply(sub)
```

```python
def sub(s):
        return (s.split(" ")[0])
train["Total_Stops"] = train["Total_Stops"].apply(sub)
train["Total_Stops"] = train["Total_Stops"].replace(to_replace = 'non-stop', value = 0)
train["Total_Stops"] = train["Total_Stops"].astype(int)
test["Total_Stops"] = test["Total_Stops"].apply(sub)
test["Total_Stops"] = test["Total_Stops"].replace(to_replace = 'non-stop', value = 0)
test["Total_Stops"] = test["Total_Stops"].astype(int)
```

```python
sns.barplot(x = train["Total_Stops"], y = train["Price"], data = train)
"Corr: " + str(train["Price"].corr(train["Total_Stops"]))
```

```python
def sub(s):
    tup = s.split("/")
    today = datetime.datetime(int(tup[-1]), int(tup[-2]), int(tup[-3]))
    return today.weekday()
train["Week_Day"] = train["Date_of_Journey"].apply(sub)
test["Week_Day"] = test["Date_of_Journey"].apply(sub)
```

```python
sns.barplot(x = train["Week_Day"], y = train["Price"], data = train)
#plt.xlabels(["Mon", "Tues","Wed","Thus","Fri", "Sat","Sun"])
```

```python
def sub(s):
        if (int(s.split(":")[0])<6):
            return 0
        elif (int(s.split(":")[0])<12):
            return 1
        elif (int(s.split(":")[0])<18):
            return 2
        else:
            return 3
train["Dep_Time"] = train["Dep_Time"].apply(sub)
train["Arrival_Time"] = train["Arrival_Time"].apply(sub)
test["Dep_Time"] = test["Dep_Time"].apply(sub)
test["Arrival_Time"] = test["Arrival_Time"].apply(sub)
```

```python
fig, axes = plt.subplots(1, 2)
sns.barplot(ax=axes[0], x = train["Dep_Time"], y = train["Price"], data = train)
sns.barplot(ax=axes[1], x = train["Arrival_Time"], y = train["Price"], data = train)
fig.tight_layout()
```

```python
def sub(s):
    if (("m" in s)==True and ("h" in s)==True): 
        return (int(s.split("h")[0])*60)+int(s.split("h")[1].split("m")[0])
    elif (("m" in s)==True and ("h" not in s)==True):
        return (int(s.split("m")[0]))
    else:
        return (int(s.split("h")[0])*60)
train["Duration"] = train["Duration"].apply(sub)
train["Duration"] = train["Duration"].astype(int)
test["Duration"] = test["Duration"].apply(sub)
test["Duration"] = test["Duration"].astype(int)
```

```python
def sub(s):
    return (int(s.split("/")[0]))
train["Day"] = train["Date_of_Journey"].apply(sub)
test["Day"] = test["Date_of_Journey"].apply(sub)
```

```python
def sub(s):
    return (int(s.split("/")[1]))
train["Month"] = train["Date_of_Journey"].apply(sub)
test["Month"] = test["Date_of_Journey"].apply(sub)
```

```python
to_encode = ['Airline', 'Route']
for i in to_encode:
    train[i+"_encoded"] = LabelEncoder().fit_transform(train[i])
    del train[i]
```

```python
for i in to_encode:
    test[i+"_encoded"] = LabelEncoder().fit_transform(test[i])
    del test[i]
```

```python
z_scores = stats.zscore(train)

abs_z_scores = np.abs(z_scores).all(axis = 1)
filtered_entries = (abs_z_scores < 3)
train = train[filtered_entries]
```

```python
train.columns
```

```python
X_train,y_train = train[['Source', 'Destination', 'Dep_Time', 'Arrival_Time', 'Duration',
       'Total_Stops', 'Week_Day', 'Day', 'Month', 'Airline_encoded',
       'Route_encoded']],train['Price']
X_test,y_test = test[['Source', 'Destination', 'Dep_Time', 'Arrival_Time', 'Duration',
       'Total_Stops', 'Week_Day', 'Day', 'Month', 'Airline_encoded',
       'Route_encoded']],test['Price']
```

```python
linear_regression = LinearRegression()
model = linear_regression.fit(X_train, y_train)
```

```python
score = model.score(X_train, y_train)
print('train accuracy : ', score)
```

```python
y_pred = model.predict(X_test)
```

```python
test_score = r2_score(y_test, y_pred)
print("test accuracy : ", test_score)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```
